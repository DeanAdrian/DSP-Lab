###Mp3 & M4a files

This subsection has a single example for each mp3 and m4a file extensions respectively. Both file extensions did not have the desired output and only provided a similar error "file does not start with RIFF id".
$$$$
RIFF (Resource Interchange File Format) is described as a general file container for storing the data in a file through tagged chunks, and WAV files are stored entirely within this format and also translates to lossless compression. However mp3 and m4a files are generally stored as a string of binary numbers which results in some data being loss in order to reduce the file size. This difference in how the files are stored is what results in the aforementioned error when attempting to output the mp3 and m4a files.

```python
sample_mp3 = thinkdsp.read_wave('sample1.mp3')

sample_mp3.plot()
plt.show()

sample_mp3.play()
IPython.display.Audio('sound.wav')
```

```python
sample_m4a = thinkdsp.read_wave('sample1.m4a')

sample_m4a.plot()
plt.show()

sample_m4a.play()
IPython.display.Audio('sound.wav')
```

#Images

This section contains the 2-dimensional signals. There are three subsections,of JPEGs, PNGs and GIF, imported with the use of the openCV library.

###JPEGs

This subsection contains two jpeg files plotted with the use of both openCV library and matplotlib. The importing of the images follows the same three steps, first is the use of open cv to read the image first and then converting the color coding of the image from openCVs default of BGR to RGB in order to keep the color consistent with the actual look of the image. and then using matplotlib to plot out the image.

In a cell below the image is a numerical representation for the RGB combination of each part of the image file.

```python
jpg_trees = cv2.imread("trees.jpg")
jpg_trees = cv2.cvtColor(jpg_trees, cv2.COLOR_BGR2RGB)
plt.imshow(jpg_trees)
```

```python
int_jpg_trees = (jpg_trees*255).astype(int)
int_jpg_trees
```

```python
jpg_gen1 = cv2.imread("gen1.jpeg")
jpg_gen1 = cv2.cvtColor(jpg_gen1, cv2.COLOR_BGR2RGB)
plt.imshow(jpg_gen1)
```

```python
int_jpg_gen1 = (jpg_gen1*255).astype(int)
int_jpg_gen1
```
