####1. Function For both left and right-hand side equations. 

```python
class Equations():
  def __init__(self,x=np.linspace(-10,10,50),z=0,y=5):
    self.x = np.linspace(-10,10,50)
    self.z = self.x + y*1j

  def leftEquation(self,y=5):
    ans = np.conj(self.z)
    plt.plot(ans)
    return ans

  def rightEquation(self,y=5):
    ans = 1/self.z
    plt.plot(self.z)
    return ans
```

```python
EQ = Equations()
```

```python
EQ.leftEquation()
```

```python
EQ.rightEquation()
```

```python
EQ.leftEquation() == EQ.rightEquation()
```

#### 2. Function for both the original and simplified versions of x[n].

```python
def plotXn(theta): 
  x = np.linspace(-10,10, 50)
  xnOrig = ((math.e ** (math.pi/2 * 1j) + math.e ** (math.pi/2 * -1j))/2) - ((math.e**1j*x*theta) - (math.e**-1j*x*theta))/1j 
  xnSimp = 1j*((math.e**1j*theta*x)-(math.e**-1j*theta*x))
  plt.plot(xnOrig, color="red", alpha= 0.45, lw=5)
  plt.plot(xnSimp, color="blue", alpha = 0.45, lw=5)
  plt.legend(["x[n]","x[n]simplified"])
  return xnOrig
```

```python
plotXn(30)
```

#### 3. Plotting the signals for (a)-(c) and marking their periods.

```python
def plotPeriod(omega):
  x = np.linspace(-5,5,50)
  y = np.exp(omega*1j*2*np.pi*x)
  plt.plot(x,y, marker='o', color='b')
  plt.show()
```

#### 4. Function for the moving average and plotting the signal for both x[n] and y[n].

```python
def moving_average(list):
  i = 0
  mov_average = []
  while i < len(list) - 1:
    average = round(np.sum(list[i:i+2]))/2
    mov_average.append(average)
    i += 1
  plt.title("Moving Average")
  plt.plot(mov_average)
  plt.plot(list)
  plt.legend(['Moving Average', 'Original'])
  return mov_average
```

```python
Xn = [0,1,2,3,0]
Yn = moving_average(Xn)
Yn
```

#### 5. Function for the determined input-output relationship.




```python
def inp_out_rel(X_n, a, b, c):
  i = 0
  output = []
  while i < len(X_n):
    yn = b * (a + X_n[i]-1) - (X_n[i]-3*(c*X_n[i] + X_n[i]))
    output.append(yn)
    i += 1
  plt.plot(X_n)
  plt.plot(output)
  plt.legend(["xn", "yn"])
  return output
```

```python
list = [1,2,3]
inp_out_rel(list, 2, 3, 5)
```
